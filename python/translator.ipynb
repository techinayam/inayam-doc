```python
import os
import json
from googletrans import Translator 
from concurrent.futures import ThreadPoolExecutor

def translate_json_files(input_folder, output_folder, target_language):
    os.makedirs(output_folder, exist_ok=True)
    translator = Translator()
    def translate_text(text, filename,  chunk_size = 2000):
        try:
            # translated = translator.translate(text, dest=target_language)
            # return translated.text
            chunks = [text[i:i+chunk_size] for i in range(0, len(text), chunk_size)]
            translated_chunks = [translator.translate(chunk, dest=target_language).text for chunk in chunks]
            return " ".join(translated_chunks)
        except Exception as e:
            print(f"Error translating '{target_language}' '{filename}': {e}")

    def translate_file(filename):
        input_path = os.path.join(input_folder, filename)
        output_path = os.path.join(output_folder, filename)
        with open(input_path, 'r', encoding='utf-8') as f:
            data = f.read()
        print(f"Translating {filename}")
        translated_data = translate_text(data, filename)
        with open(output_path, 'w', encoding='utf-8') as f:
            f.write(translated_data)
        print(f"{filename}Translated and saved: {output_path}")

    files_to_translate = [filename for filename in os.listdir(input_folder) if filename.endswith(".md")]
    
    with ThreadPoolExecutor(max_workers=10) as executor:
        executor.map(translate_file, files_to_translate)
#    locales: ['en', 'ta', 'hi', 'es', 'de', 'ru', 'ja', 'fr', 'pt', 'zh', 'ar'],

input_folder = "/Users/rakesh.c/rakesh/my/am/inayam/inayam-doc/utilities/en"
output_folder = "/Users/rakesh.c/rakesh/my/am/inayam/inayam-doc/utilities/de"
target_language = "ta" 

translate_json_files(input_folder, output_folder, target_language)
```

```python
#outside
locales = ['ta', 'hi', 'es', 'de', 'ru', 'ja', 'fr', 'pt', 'zh-CN', 'ar', 'it', 'ko']

base_folder = "/Users/rakesh.c/rakesh/my/am/inayam/inayam-doc/utilities/"
for locale in locales:
    print(f'\n=== Started processing {locale} ===')
    input_folder = os.path.join(base_folder, 'en')
    utility = 'zh' if locale == 'zh-CN' else locale
    output_folder = os.path.join(base_folder, utility)
    translate_json_files(input_folder, output_folder, locale)
    print(f'=== Finished processing {locale} ===\n')
```
